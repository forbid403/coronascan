```python
import openpyxl
import pandas as pd
import json
import yaml
```

```python
filename = "test.xlsx"
book = openpyxl.load_workbook(filename)
sheets = book.worksheets
```

```python
dataf = pd.DataFrame()
data = []
delete_index = [8, 2, 3, 3, 3]
#0, 1 -> 입국 금지, 2, 3, 4 -> 입국 제한

for sheetNum, sheet in enumerate(sheets) :
    for idx, row in enumerate(sheet.rows):
        data.append([row[1].value, row[2].value])
    del data[:delete_index[sheetNum]]
```

```python
print(len(dataf))
print(len(data))
```

```python
for index, row in enumerate(data) :
    if row[0] == None or row[1] == '조  치  사  항' :
        del data[index]
        continue
    dataf.loc[index, 'nation_kr'] = row[0]
    dataf.loc[index, 'detail'] = row[1]
dataf = dataf.reset_index(drop = True)
```

```python
filename = "ISO국가코드.xlsx"
book = openpyxl.load_workbook(filename)
sheet = book.worksheets[0]
countries = []
for row in sheet.rows:
    countries.append([row[3].value, row[1].value])
del countries[:4]

countries = dict(countries)
```

```python
for index, row in enumerate(data) :
    if row[0] in countries :
        dataf.loc[index, 'nation_eng'] = countries[row[0]]
        dataf.loc[index, 'marker'] = False
    else :
        dataf.loc[index, 'marker'] = True
```

```python
import openpyxl
import pandas as pd
import json
import yaml
```

```python
filename = "test.xlsx"
book = openpyxl.load_workbook(filename)
sheet = book.worksheets[0]
```

```python
data = []
for row in sheet.rows:
    data.append([row[1].value, row[2].value])
del data[:8]
```

```python
dataf = pd.DataFrame()
for index, row in enumerate(data) :
    if row[0] == None or row[1] == '조  치  사  항' :
        del data[index]
        continue
    dataf.at[index, 'nation_kr'] = row[0]
    dataf.at[index, 'detail'] = row[1]
dataf = dataf.reset_index(drop = True)
```

```python
filename = "ISO국가코드.xlsx"
book = openpyxl.load_workbook(filename)
sheet = book.worksheets[0]
countries = []
for row in sheet.rows:
    countries.append([row[3].value, row[1].value])
del countries[:4]

countries = dict(countries)
```

```python
for index, row in enumerate(data) :
    if row[0] in countries : 
        dataf.at[index, 'nation_eng'] = countries[row[0]]
        dataf.at[index, 'marker'] = False
    else:
        dataf.at[index, 'marker'] = True

dataf
```
